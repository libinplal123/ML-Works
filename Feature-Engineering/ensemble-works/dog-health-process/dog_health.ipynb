```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\libin_urv2w13\\Desktop\\mljourney\\Feature-Engineering\\ensemble-works\\dog-health-process\\dog_health_dataset_10000.csv.xls")
```

```python
df.drop(columns=["Precautions","Treatment","Emergency"],inplace=True)
df.columns
```

```python
symptom_columns = ["Symptom 1","Symptom 2","Symptom 3","Symptom 4","Symptom 5"]
df["All_symptoms"] = df[symptom_columns].values.tolist()
df.drop(columns=symptom_columns)
```

```python
df.columns
```

```python
# applying multilabelbinarizer
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
symptom_encoded = mlb.fit_transform(df["All_symptoms"])
drop_cols = symptom_columns.copy()
drop_cols.append("All_symptoms")
```

```python
df.drop(columns=drop_cols,inplace=True)
```

```python
print(df.head(3))
symptom_encoded.dtype
```

```python
symptom_encoded_df = pd.DataFrame(symptom_encoded,columns=mlb.classes_,index=df.index)
```

```python
df = pd.concat([df,symptom_encoded_df],axis=1)
```

```python
df.head()
```

```python
disease = df["Disease"]
print(disease)
df.drop(columns="Disease",inplace=True)
df.columns
```

```python
df["Disease"] = disease
df.columns
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x 
```

```python
df["Dog Breed"].value_counts()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([("encoder",OneHotEncoder(),["Dog Breed"])],remainder="passthrough")
x = pd.DataFrame(ct.fit_transform(x))
x.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_train
```

```python
from sklearn.naive_bayes import GaussianNB
gb = GaussianNB()
gb.fit(x_train,y_train)
```

```python
y_pred = gb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
models = [DecisionTreeClassifier,KNeighborsClassifier,GaussianNB,RandomForestClassifier]
for model in models:
    md = model()
    md.fit(x_train,y_train)
    y_pred = md.predict(x_test)
    print("======",model,"=====")
    print("accuracy_score",accuracy_score(y_test,y_pred)*100)
    print("precision_score",precision_score(y_test,y_pred,average="weighted",zero_division=0)*100)
    print("f1_score",f1_score(y_test,y_pred,average="weighted",zero_division=0)*100)
    print("recall_score",recall_score(y_test,y_pred,average="weighted",zero_division=0)*100)
```

```python
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier
```

```python
abc = AdaBoostClassifier()
abc.fit(x_train,y_train)
```

```python
y_pred = abc.predict(x_test)
from sklearn.metrics import accuracy_score
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
```

```python
gbc = GradientBoostingClassifier()
gbc.fit(x_train,y_train)
```

```python
y_pred = gbc.predict(x_test)
from sklearn.metrics import accuracy_score
print("accuracy_score",accuracy_score(y_test,y_pred)*100)
```
