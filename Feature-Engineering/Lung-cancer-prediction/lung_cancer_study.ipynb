```python
# spliting into two columns
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
# checking null values and filling with imputer
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy="mean")
X.iloc[:,1:14]=imputer.fit_transform(X.iloc[:,1:14])
```

```python
# Encoding categorical values in X using one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([("encoder",OneHotEncoder(),[0])],remainder="passthrough") # [0] represents column to transform
X=pd.DataFrame(ct.fit_transform(X))
X.head() # 0,1 represent M,F
```

```python
# encoding categorical values of Y using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
Y
```

```python
# spliting X and Y into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1) # random state: to select same training values for each run
X_train
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.fit(X_test)
X_train
```
