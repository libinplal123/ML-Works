```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
print(x.columns)
print(y)
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values="?",strategy="most_frequent")
x.iloc[:,:]=imputer.fit_transform(x.iloc[:,:])
```

```python
x.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer([("encoder",OneHotEncoder(sparse_output=False),["workclass","marital.status","occupation","relationship","race","sex","native.country"])],remainder="passthrough")
x = pd.DataFrame(ct.fit_transform(x),columns=ct.get_feature_names_out())
x
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
x_train,y_train = sm.fit_resample(x_train,y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred = knn.predict(x_test)
from sklearn.metrics import accuracy_score,recall_score,f1_score
print(accuracy_score(y_test,y_pred)*100)
print(recall_score(y_test,y_pred)*100)
print(f1_score(y_test,y_pred)*100)
```
