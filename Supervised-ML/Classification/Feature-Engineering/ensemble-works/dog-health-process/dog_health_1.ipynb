```python
df.drop(columns=["Precautions","Treatment","Emergency"],inplace=True)
df.columns
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
x
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("encoder",OneHotEncoder(sparse_output=False),['Dog Breed', 'Symptom 1', 'Symptom 2', 'Symptom 3', 'Symptom 4', 'Symptom 5'])],remainder="passthrough")
x_transformed = ct.fit_transform(x)

# 3. Re-create the DataFrame with the correct column names
column_names = ct.get_feature_names_out()
x = pd.DataFrame(x_transformed, columns=column_names)

# Now check the result
x.head()
```

```python
print(x.columns.tolist())
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
X_train.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
columns = ['remainder__Dog Age', 'remainder__Dog Weight']
X_train[columns] = scaler.fit_transform(X_train[columns])
X_test[columns] = scaler.transform(X_test[columns])
X_train[columns]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("--- MODEL PERFORMANCE ---")
print(f"Overall Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")

print("\nDetailed Classification Report:")
# Use le.classes_ to show the actual disease names instead of numbers!
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.ensemble import RandomForestClassifier

# 1. Initialize Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# 2. Train
rf_model.fit(X_train, y_train)

# 3. Check Accuracy
rf_preds = rf_model.predict(X_test)
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_preds) * 100:.2f}%")
```
