```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\libin_urv2w13\\Desktop\\mljourney\\Supervised-ML\\Simple-Linear-Reg\\tv-mraketing-process\\tvmarketing.csv.xls")
```

```python
sns.regplot(x=df["TV"],y=df["Sales"])
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.coef_)      # m, slope
print(lr.intercept_) # y, intercept
```

```python
# y for a value of x
x.head(1)
```

```python
# y for x 230.1 = 0.0481*230 + 6.865 = 17.92, actual value 22.1 error = 4 
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
print(r2_score(y_test,y_pred)*100)
print(mean_squared_error(y_test,y_pred)*100)
print(mean_absolute_error(y_test,y_pred)*100)
print(root_mean_squared_error(y_test,y_pred)*100)
```
