```python
df["mileage"] = df["mileage"].str.split(" ").str[0]
df["mileage"] = df["mileage"].astype(float)
df.info()
```

```python
df["engine"] = df["engine"].str.split(" ").str[0]
df["engine"] = df["engine"].astype(float)
df.info()
```

```python
df["max_power"] = df["max_power"].str.split("b").str[0]
df["max_power"] = df["max_power"].replace(' ',np.nan)
df["max_power"] = df["max_power"].astype(float)
```

```python
# df["torque"] = df["torque"].replace(' ',np.nan)
df["torque"] = df["torque"].str.strip().str.split("N|n|@|k|\(| ").str[0]
df["torque"] = df["torque"].astype(float)
df.torque
```

```python
x = df.iloc[:,:]
x = x.drop(columns="selling_price")
x.head()
```

```python
y = df.iloc[:,1]
y
```

```python
x.columns[6:11]
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan,strategy="most_frequent")
x.iloc[:,6:11] = imputer.fit_transform(x.iloc[:,6:11])
```

```python
x.head()
```

```python
x.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("encoder",OneHotEncoder(sparse_output=False),["brand","fuel","seller_type","transmission","owner"])],remainder="passthrough")
x = pd.DataFrame(ct.fit_transform(x),columns=ct.get_feature_names_out())
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=4)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,root_mean_squared_error
print(r2_score(y_test,y_pred)*100)
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
```

```python
models = [DecisionTreeRegressor,RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor]
for model in models:
    md = model()
    md.fit(x_train,y_train)
    y_pred = md.predict(x_test)
    print("=====",model,"====")
    print(r2_score(y_test,y_pred)*100)
```
