```python
import pandas as pd
import seaborn as sns
df = pd.read_excel("C:\\Users\\libin_urv2w13\\Desktop\\mljourney\\seaborn-works\\case-study-flights\\Data_Train.xlsx")
print(df.head()) # pip install openpyxl for reading excel files
```

```python
df.isnull().sum()
```

```python
# cleaning date
df["day"]=df["Date_of_Journey"].str.split("/").str[0]
df["month"]=df["Date_of_Journey"].str.split("/").str[1]
df["year"]=df["Date_of_Journey"].str.split("/").str[2]
# deleting the "Date_of_Journey" column
df.drop(columns="Date_of_Journey",inplace=True)
print(df.columns)
```

```python
df["day"]=df["day"].astype("int")
df["month"]=df["day"].astype("int")
df["year"]=df["day"].astype("int")
df.info()
```

```python
df["departure_hour"]=df["Dep_Time"].str.split(":").str[0]
df["departure_minute"]=df["Dep_Time"].str.split(":").str[1]
df.drop(columns="Dep_Time",inplace=True)
print(df.columns)
```

```python
df["Arrival_Time"] = df["Arrival_Time"].str.split(" ").str[0]
df["Arrival_hr"] = df["Arrival_Time"].str.split(":").str[0]
df["Arrival_min"] = df["Arrival_Time"].str.split(":").str[1]
df.drop(columns="Arrival_Time",inplace=True)
print(df.columns)
```

```python
print(df["Total_Stops"].value_counts())
print(df[df["Total_Stops"].isnull()])
```

```python
import numpy as np
print(df["Total_Stops"].unique())
df["Total_Stops"]=df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4,np.nan:1})
```

```python
print(df["Total_Stops"])
print(df["Total_Stops"].value_counts())
```

```python
df["Duration_hr"]=df["Duration"].str.split(" ").str[0]
df["Duration_hr"]=df["Duration_hr"].str.rstrip("h")
df["Duration_min"]=df["Duration"].str.split(" ").str[1]
df["Duration_min"]=df["Duration_min"].str.rstrip("m")
# print(df["Duration_min"].value_counts(dropna=False))
df.fillna(df["Duration_min"].mode()[0],inplace=True )
print(df["Duration_min"].value_counts(dropna=False))
```
