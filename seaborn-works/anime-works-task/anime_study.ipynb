```python
import seaborn as sns
import pandas as pd
df = pd.read_csv("C:\\Users\\libin_urv2w13\\Desktop\\mljourney\\seaborn-works\\anime-works-task\\anime.csv")
```

```python
print(df.shape)
print(df.columns)
print(df.isnull().sum())
```

```python
df["synopsis"]=df["synopsis"].fillna(df["synopsis"].mode()[0],inplace=True)
df["start_date"]=df["start_date"].fillna(df["start_date"].mode()[0],inplace=True)
df["end_date"]=df["end_date"].fillna(df["end_date"].mode()[0],inplace=True)
df["episodes"]=df["episodes"].fillna(df["episodes"].mode()[0],inplace=True)
```

```python
print(df.isnull().sum())
```

```python
sns.lineplot(data=df,x="score",y="rank")
```
