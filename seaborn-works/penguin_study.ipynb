```python
import seaborn as sns
sns.get_dataset_names()
```

```python
data = sns.load_dataset("penguins")
print(data.head())
print(data.columns)
print(data.isnull().sum())
```

```python
data["species"].value_counts()
```

```python
data.shape
```

```python
sns.scatterplot(data=data,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.scatterplot(data=data,x="flipper_length_mm",y="body_mass_g",hue="island")
```

```python
sns.scatterplot(data=data,x="bill_length_mm",y="body_mass_g",hue="island",palette="bright")
sns.despine(left=True)
sns.set_style("darkgrid")
sns.set_context("notebook")
```

```python
sns.stripplot(data=data,x="island",y="body_mass_g")
```

```python
sns.stripplot(data=data,x="species",y="body_mass_g")
```

```python
sns.stripplot(data=data,x="species",y="body_mass_g",hue="island")
```

```python
sns.histplot(data=data,x="body_mass_g",hue="sex")
```

```python
sns.regplot(data=data,x="flipper_length_mm",y="body_mass_g",color="purple") # regression line
```

```python
sns.lineplot(data=data,x="flipper_length_mm",y="body_mass_g")
```

```python
sns.lineplot(data=data,x="flipper_length_mm",y="body_mass_g",hue="sex")
```

```python
sns.lineplot(data=data,x="flipper_length_mm",y="body_mass_g",hue="island")
```

```python
sns.barplot(data=data,x="island",y="body_mass_g")
sns.set_style("whitegrid")
```

```python
import pandas as pd
data.groupby("island")["body_mass_g"].mean() # by default body mass calculated as avg
```

```python
import numpy as np
sns.barplot(data=data,x="island",y="body_mass_g",estimator=np.sum) # use estimator to change default avg value
sns.set_style("whitegrid")
```

```python
sns.boxplot(data=data,x="species",y="body_mass_g",hue="sex")
```

```python
sns.violinplot(data=data,x="species",y="body_mass_g")
```

```python
sns.violinplot(data=data,x="species",y="body_mass_g",hue="sex",split=True)
```

```python
# KDE (kernel density estimator)
sns.kdeplot(data=data,x="body_mass_g",fill=True,hue="sex")
```
