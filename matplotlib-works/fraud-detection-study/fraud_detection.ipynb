```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\libin_urv2w13\\Desktop\\mljourney\\matplotlib-works\\fraud-detection-study\\Fraud Detection Transactions Dataset.csv")
```

```python
print(df.shape)
print(df.columns)
```

```python
df.isnull().sum()
```

```python
# all transaction types
df["Transaction_Type"].unique()
```

```python
# average transaction amount per transaction type
df.groupby("Transaction_Type")["Transaction_Amount"].mean()
```

```python
# most frequent transaction type
df["Transaction_Type"].mode()[0]
```

```python
# all unique locations
df["Location"].unique()
```

```python
# device type and transaction amt plot
device_transaction_amnt = df.groupby("Device_Type")["Transaction_Amount"].mean()
plt.bar(device_transaction_amnt.index,device_transaction_amnt.values)
```

```python
# merchant category and transaction amount
merc_transaction_amnt=df.groupby("Merchant_Category")["Transaction_Amount"].mean()
plt.bar(merc_transaction_amnt.index,merc_transaction_amnt.values)
```

```python
# average transaction amount and card type
avg_amt_card_type=df.groupby("Card_Type")["Avg_Transaction_Amount_7d"].mean()
plt.bar(avg_amt_card_type.index,avg_amt_card_type.values)
```
