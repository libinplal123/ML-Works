```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\libin_urv2w13\\Desktop\\mljourney\\matplotlib-works\\company-sales-study\\company_sales_data.csv")
```

```python
print(df.shape)
print(df.columns)
print(df.isnull().sum())
```

```python
# total profit
plt.plot(df.groupby("month_number")["total_profit"].sum(),linestyle="dotted",color="red",marker="o",linewidth=3, label="month,units sold")
plt.legend()
plt.xlabel("Month Number")
plt.ylabel("Sold units number")
```

```python
plt.plot(df["month_number"],df["facecream"],marker="s",markerfacecolor="green",label="facecream")
plt.plot(df["month_number"],df["facewash"],marker="s",markerfacecolor="green",label="facewash")
plt.plot(df["month_number"],df["toothpaste"],marker="s",markerfacecolor="green",label="toothpaste")
plt.plot(df["month_number"],df["bathingsoap"],marker="s",markerfacecolor="green",label="bathingsoap")
plt.plot(df["month_number"],df["shampoo"],marker="s",markerfacecolor="green",label="shampoo")
plt.plot(df["month_number"],df["moisturizer"],marker="s",markerfacecolor="green",label="moisturizer")
plt.legend()
```

```python
plt.scatter(df["month_number"],df["toothpaste"])
```

```python
paste_sales=df.groupby("month_number")["facecream"].mean()
facewash_sales=df.groupby("month_number")["facewash"].mean()
plt.bar(paste_sales.index,paste_sales.values,label="facecream")
plt.bar(facewash_sales.index,facewash_sales.values,label="facewash")
plt.legend()
plt.grid()
```

```python
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
sales_summary={p:df[p].sum() for p in products}
plt.pie(sales_summary.values(),labels=products,autopct="%1.1f%%")
```
