```python
import matplotlib.pyplot as plt
food_logs=["dosa","tea","meals","biriyani"]
calories=[100,20,150,200]
plt.plot(food_logs,calories)
plt.show()
```

```python
courses=["django","DS","AI","ML"]
student_count_jan=[10,12,15,20] # can use color picker for custom colors
student_count_feb=[7,8,11,15] 
plt.plot(courses,student_count_jan,marker="^",linestyle="dotted",color="green",linewidth=2)
plt.plot(courses,student_count_feb,marker=".",linestyle="dashdot",color="red",linewidth=2)
plt.title("Luminar TSR",color="green",size=20,weight="bold")
plt.xlabel("courses",color="red")
plt.ylabel("student_count",color="green")
plt.show()
```
